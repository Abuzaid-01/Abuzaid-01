```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
import pandas as pd

# This path is created automatically by Kaggle when you add the dataset
df = pd.read_csv("/kaggle/input/gender-by-name/name_gender_dataset.csv")

# Display the first few rows
df.head()
```

```python
df.isna().sum()
```

```python
df.Name = df.Name.str.lower()
```

```python
df
```

```python
df["label"] = df["Gender"].map({"M" : 1, "F" : 0 })
```

```python
df
```

```python
vectorizer = CountVectorizer(analyzer='char', ngram_range=(2, 4))
X = vectorizer.fit_transform(df["Name"])
y = df["label"]

# from sklearn.ensemble import RandomForestClassifier

# model = RandomForestClassifier()
# model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2,stratify = y, random_state=42
)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Step 8: Test Your Own Names
def predict_gender(name):
    name = [name.lower()]
    name_vec = vectorizer.transform(name)
    pred = model.predict(name_vec)[0]
    return "Male" if pred == 1 else "Female"

# Try it out!
print(predict_gender("yusha"))
print(predict_gender("Sophia"))
print(predict_gender("Zaid"))
print(predict_gender("Ritika"))
```
